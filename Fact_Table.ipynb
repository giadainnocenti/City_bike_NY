# Fact Table
This workbook is just use to read all the CSV files contained in Resources20-21 (containing the last 6 months of published data) and merge them in one big table that I am going to use within Tableau. 

```python
import pandas as pd
import os
import fnmatch
```

```python
os.chdir('./Resources20-21/')
```

```python
data =[]
for file in os.listdir('./'):
    if fnmatch.fnmatch(file, '*.csv'):
        data += [pd.read_csv(file)] 
df = pd.concat(data, join='outer', axis=0)
df
```

```python
#check if there is any NA value
df.isna().sum()
```

```python
# checking for wrong values
df['birth year'].unique()
```

```python
#if the birth year is lower than 1940 the customer/subscribes is more than 79 years old in 2019.
# I am assuming those are wrong entries and as a consequence I am dropping those rows.
df = df[df['birth year'] >= 1940]
df['birth year'].unique()
```

```python
#I am dropping the rows of people with unknown gender, just because I need to trim the dataset under 15M rows to handle it with Tableau Public
df = df[df['gender'] != 0]
df
```

```python
#replacing the numerical number for sex with characters
df["gender"].replace({1: "male", 2: "female"}, inplace=True)
df
```

```python
# I am converting the trip time in minutes instead of seconds
df['tripduration']=df['tripduration'].divide(60)
df
```
